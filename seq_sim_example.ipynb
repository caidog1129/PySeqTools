```python
import numpy as np
import pandas as pd
import networkx as nx
from matplotlib import pyplot as plt
import random
from PySeqTools import PySeqTools as seq
```

```python
G = nx.gnp_random_graph(12, .06, directed=True,seed=1)
nx.draw(G)
```

Suppose we want to analysis a sequence [11, 9, 10, 4, 11, 9], with sum distance 1.1435.

```python
df = seq.random_walk(G,6,10000)
```

```python
df["sum_"] = df["2"].apply(sum)
```

```python
df["sum_"].plot(kind = "hist")
```
